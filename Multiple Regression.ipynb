## Import the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Read the dataset

```python
df=pd.read_csv('mtcars.csv')
df
```

### Separate the input and output variables

```python
#disp, hp and wt ->input variables
x=df[['disp', 'hp', 'wt']]
#mpg-> output variable
y=df['mpg']
```

```python
x.shape
```

### Build the model

```python
from sklearn.linear_model import LinearRegression
```

```python
reg =  LinearRegression()
```

```python
reg.fit(x, y)
```

```python
disp = 221
hp = 102
wt = 2.91 #what is mpg of the car
```

```python
reg.predict([[disp, hp, wt]])
```

```python
reg.score(x, y) ##Accuracy
```

```python
#positive linear and negative linear 
plt.figure(figsize=(10,10))
plt.subplot(2, 2, 1)
plt.xlabel('disp')
plt.ylabel('mpg')
plt.scatter(df['disp'], y)

plt.subplot(2, 2, 2)
plt.xlabel('hp')
plt.ylabel('mpg')
plt.scatter(df['hp'], y)

plt.subplot(2, 2, 3)
plt.xlabel('wt')
plt.ylabel('mpg')
plt.scatter(df['wt'], y);
```
