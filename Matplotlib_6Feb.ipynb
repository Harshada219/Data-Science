```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([12, 23, 34, 45, 56])
y = np.array([15, 21, 37, 42, 59])
z = np.array([87, 75, 54, 36, 25])
```

## Lineplot

```python
plt.title('Line plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.plot(x, y); #; ->to hide the unneccessary lines
```

```python
plt.title('Line plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.plot(x, z, color= 'y', marker ='o', mfc = 'white', ls ='--',lw =5, ms=10, label='2023');

plt.xlabel('First value')
plt.ylabel('Second value')
plt.plot(x, y, color= 'r', marker ='D', mfc = 'blue', ls ='--',lw =5, ms=10, label='2024');
plt.legend();
```

```python
plt.title('Line plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.plot(x, y, color= 'r', marker ='D', mfc = 'blue', ls ='--',lw =5, ms=25);
#Marker=> >, <, o, ^, D, V
#Marker font color
#ls-> Line-space
#lw->Line-width
#ms-> Marker-size
```

```python
plt.title('Line plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.plot(x, z, color= 'y', marker ='o', mfc = 'white', ls ='--',lw =5, ms=10, label='2023');
plt.plot(x, y, color= 'r', marker ='*', mfc = 'blue', ls ='--',lw =5, ms=15, label='2024');
plt.legend(); #to print labels
plt.grid() #Give graph lines
```

## Subplot

```python
plt.figure(figsize=(16, 9))
plt.subplot(2, 2, 1)
plt.title('Line plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.plot(x, z, color= 'y', marker ='o', mfc = 'white', ls ='--',lw =5, ms=10, label='2023');
plt.legend();
plt.grid()
           
           
plt.subplot(2, 2, 2)
plt.title('Line plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.plot(x, y, color= 'r', marker ='*', mfc = 'blue', ls ='--',lw =5, ms=15, label='2024');
plt.legend(); #to print labels
plt.grid() #Give graph lines
```

## Bar plot

```python
courses=['Java', 'C', 'Python', 'R', 'Kotlin']
students =[67, 70, 74, 45, 50]
```

```python
plt.title('Student Enrollment')
plt.xlabel('Courses')
plt.ylabel('No. of Students')
plt.grid()
plt.bar(courses, students, color=['yellow','red','blue','cyan','orange'], width=0.5, edgecolor ='green', linewidth =4, ls='--', alpha = 0.8);
#alpha->how dark or faint the graph
```

# Scatter plot 

```python
plt.title('Line plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.scatter(x, y, marker ='o');
```

## Box Plot

```python
#For visualize the outliers we use box plot
marks=[67, 63, 62, 69, 59, 66, 61, 70, 80]
```

```python
plt.boxplot(marks);
#Median -> middle line
#rectangle -> intre-quartile edge(iqr)
```

```python
Q3 = np.percentile(marks, 75)
Q3
```

```python
Q1 = np.percentile(marks, 25)
Q1
```

```python
IQR = Q3 - Q1
IQR
```

```python
Upper_bound = Q3 + 1.5 * IQR
Upper_bound
```

```python
Lower_bound = Q1 - 1.5 * IQR
Lower_bound
```
