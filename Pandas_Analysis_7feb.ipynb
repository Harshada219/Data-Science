## Merging of data frame

```python
# Create data frame containing => roll, cgpa, gender, where roll is same as df
gender=['m', 'm', 'm', 'm', 'f', 'f', 'm', 'f', 'm', 'm']
cgpa = df['marks']/10+0.5
roll = range(101, 111)
```

```python
newdf = pd.DataFrame({
    'roll' :roll,
    'gender':gender,
    'cgpa':cgpa
})
newdf
```

```python
pd.merge(df, newdf, on='roll')
df
```
